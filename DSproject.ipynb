```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('zarplata3.csv')
df
```

```python
# строим график для добычи полезных ископаемых
x1 = df['год'].tolist()
y1 = df['добыча полезных ископаемых'].tolist()

plt.plot(x1, y1)
```

```python
# строим график для строительства
x2 = df['год'].tolist()
y2 = df['строительство'].tolist()

plt.plot(x2, y2)
```

```python
# строим график для здравоохранения
x3 = df['год'].tolist()
y3 = df['здравоохранение и социальные услуги'].tolist()

plt.plot(x3, y3)
```

### Выводы

По графикам видно, что рост зарплат в отрасли "Добыча полезных ископаемых" плавный, без каких-либо особенностей, тогда как график "Строительство" демонстрирует небольшой спад в 2009 году, что, вероятнее всего, связано с кризисом 2008-2009 года, когда объемы строительства снизились из-за снижения инвестиций. На графике "Здравоохранение" подъем становится немного более крутым после 2018 года (к этому времени зарплаты врачей должны были вырасти до определенных величин, согласно указам президента от 2012 г.) 

### Пересчитываем зарплаты с учетом инфляции

Зарплаты пересчитывались по этой формуле:

*Реальная зарплата = зарплата - ((зарплата x уровень инфляции) / 100)*

```python
# инфляция за 2000 год взята равной 0, т.к. это инфляция по отношению к предыдущему, 1999 году, который не рассматривается

infl = pd.read_csv('zarplata_infl.csv')
infl
```

### Как влияет инфляция на изменение зарплаты по сравнению с предыдущим годом?

Посмотрим, как растут зарплаты в процентном соотношении

```python
dob = pd.read_csv('dobycha_infl.csv')
dob
```

```python
dob.loc[(dob['рост выше инфляции'].values == ['Нет'])]
```

Видно, что в 2009, 2014-2015 и 2021 году реальные зарплаты в добыче не росли, а сокращались, что, вероятно, связано с ценами на нефть, кризисными и санкционными явлениями.

```python
stroy = pd.read_csv('stroitelstvo_infl.csv')
stroy
```

```python
stroy.loc[(stroy['рост выше инфляции'].values == ['Нет'])]
```

Строительство сходным образом демонстрирует снижение реальных зарплат приблизительно в те же годы и по тем же причинам.

```python
zdrav = pd.read_csv('zdrav_infl.csv')
zdrav
```

```python
zdrav.loc[(zdrav['рост выше инфляции'].values == ['Нет'])]
```

Помимо послекризисных годов (2010, 2014-2016), в здравоохранении снижение реальных зарплат произошло после завершении эпидемии ковида в 2021-2022 году, когда сократились объемы дополнительных выплат за работу с ковидными больными.

### Делаем графики для сравнения зарплат без инфляции и с инфляцией

```python
# строим график для добычи полезных ископаемых без инфляции и с инфляцией
x1 = infl['год'].tolist()
y1 = infl['добыча полезных ископаемых'].tolist()

plt.plot(x1, y1)

x2 = infl['год'].tolist()
y2 = infl['добыча с инфл'].tolist()

plt.plot(x2, y2)
```

```python
# строим график для строительства без инфляции и с инфляцией
x1 = infl['год'].tolist()
y1 = infl['строительство'].tolist()

plt.plot(x1, y1)

x2 = infl['год'].tolist()
y2 = infl['строительство с инф'].tolist()

plt.plot(x2, y2)
```

```python
# строим график для здравоохранения без инфляции и с инфляцией
x1 = infl['год'].tolist()
y1 = infl['здравоохранение и социальные услуги'].tolist()

plt.plot(x1, y1)

x2 = infl['год'].tolist()
y2 = infl['здрав с инфл'].tolist()

plt.plot(x2, y2)
```

Оранжевым цветом на каждом графике указан рост зарплат с учетом инфляции. По сравнению с синими графиками наглядно видно, что реальный рост зарплат не всегда был плавным, а главное, зарплаты не всегда действительно росли. Особенно хорошо проседание видно в строительстве в 2009 и 2015 году и здравоохранении в 2021-2022 году.
